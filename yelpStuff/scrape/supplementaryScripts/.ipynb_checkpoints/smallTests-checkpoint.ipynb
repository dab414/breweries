```python
import time
import requests

def testProxy(proxies):
	'''
	takes as input proxies dict
	returns 0 if proxy successfully connects to a server, 1 otherwise
	'''
	start = time.time()

	status = 1

	while time.time() - start < 5:
		print time.time() - start
		result = requests.get('http://www.ipconfig.me/ip', proxies = proxies).text
		status = 0
		break

	if result:
		if result != proxies['http'].split(':')[0]:
			status = 1

	return status
```

```python
import urllib2, socket

def is_bad_proxy(pip):    
    try:        
        proxy_handler = urllib2.ProxyHandler({'http': pip})        
        opener = urllib2.build_opener(proxy_handler)
        opener.addheaders = [('User-agent', 'Mozilla/5.0')]
        urllib2.install_opener(opener)        
        req=urllib2.Request('http://www.ipconfig.me/ip')
        sock=urllib2.urlopen(req)
    except urllib2.HTTPError, e:        
        print 'Error code: ', e.code
        return e.code
    except Exception, detail:

        print "ERROR:", detail
        return 1
    return 0

good_pip = '104.248.152.88:3128'
bad_pip = '88.248.23.216:36426'

is_bad_proxy(good_pip)
```

```python
import signal

def handler(signum, frame):
    print 'Forever is over!'
    raise Exception()
    
def loop_forever():
    import time
    while 1:
        print 'sec'
        time.sleep(1)
        

signal.signal(signal.SIGALRM, handler)
signal.alarm(10)

try:
    loop_forever()
except Exception, exc:
    print exc
    

    
signal.alarm(0)
```

```python
import requests

def testProxy(proxy):
	'''
	takes as input proxies dict
	returns 0 if proxy successfully connects to a server, 1 otherwise
	logic of function from: https://stackoverflow.com/questions/492519/timeout-on-a-function-call
	'''  
	import signal
	status = 1



	def handler(signum, frame):
		print 'Proxy timed out'
		raise Exception()

	def attemptRequest(proxy):
		return requests.get('http://www.ipconfig.me/ip', proxies = proxy).text

	signal.signal(signal.SIGALRM, handler)

	signal.alarm(5)

	try:
		result = attemptRequest(proxy)
		if result == proxy['http'].split(':')[0]:
			print 'Proxy is fire'
			status = 0

	except:
		print 'Proxy is no good'

	return status

oldProxy = {'http': '198.27.67.35:3128'}
newProxy = {'http': '68.177.71.120:61427'}

testProxy(newProxy)
```
